## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges, which you can read more about on their websites or additional outside sources (a quick Google search will help you understand the scores for each test):
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. Opponents of these tests claim that these tests are not accurate measures of students potential or ability and serve as an inequitable barrier to entry. Lately, more and more schools are opting to drop the SAT/ACT requirement for their Fall 2021 applications ([*read more about this here*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

Despite colleges and universities putting the requirement of standardized tests (ACTs & SATs) to a halt ([source](https://www.smithsonianmag.com/innovation/has-pandemic-put-end-to-sat-act-180978167/)), taking standardized tests will still provide the students with more alternatives in terms of the selection of schools they can apply to. There are still a large number of schools with compulsory standardized tests ([source](https://www.compassprep.com/college-profiles/)), hence if they are unable to apply for schools with optional standardized test, it keeps their options for alternative universities open. 

Instead of the investing in schools to increase participation rate, we can also increase the number of students going to universities by investing in the education structure (e.g. providing subsidised tuition, extra classes, etc.). Even though most prestigious universities and colleges do not require ACTs and SATs, those which are less prestigious still have ACTs and SATs as an entrance requirement. This way, we can increases competitiveness and improve the performance of students in California, as it allows more students to get into universities.

From the above we can see that the following needs to be done:

```python
#Columns 10 to 15 (AvgScrRead, AvgScrEng, AvgScrMath, AvgScrSci, NumGE21, PctGE21) are corrupted with illegal characters
#as they are supposed to be floats.

#replace '*' with mean values of each column for Average Subject Scores.
act_ca['AvgScrRead'] = act_ca['AvgScrRead'].replace('*', np.mean(pd.to_numeric(act_ca['AvgScrRead'], errors='coerce')))
act_ca['AvgScrEng'] = act_ca['AvgScrEng'].replace('*', np.mean(pd.to_numeric(act_ca['AvgScrEng'], errors='coerce')))
act_ca['AvgScrMath'] = act_ca['AvgScrMath'].replace('*', np.mean(pd.to_numeric(act_ca['AvgScrMath'], errors='coerce')))
act_ca['AvgScrSci'] = act_ca['AvgScrSci'].replace('*', np.mean(pd.to_numeric(act_ca['AvgScrSci'], errors='coerce')))

#As this is a percentage of _______, we will use the overall State of california percentage as a average, which is 55.0%
act_ca['PctGE21'] = act_ca['PctGE21'].replace('*',55.0)

#Converting numerical strings into floats for all score related columns
act_ca[['AvgScrRead','AvgScrEng','AvgScrMath','AvgScrSci','PctGE21']] = act_ca[['AvgScrRead','AvgScrEng','AvgScrMath','AvgScrSci','PctGE21']].applymap(float).round(2)

#Replacing '*' values in NumGE21 by multiplying NumTskTakr and PctGE21
act_ca['NumGE21'] = (act_ca['NumTstTakr'] * (act_ca['PctGE21']/100)).round(2)

#Change PctGE21 percentage to decimal (4 d.p.), unable to use str_to_percent function as it does not have '%' to strip
act_ca['PctGE21'] = (act_ca['PctGE21']/100).round(4)

#Checking for all the changes done above
act_ca.head()
```

```python
#if CDS, CCode, CDCode, SCode, SName, DName is 0, sub with NaN to standardise all null datas.
act_ca.loc[act_ca['CDS']==0, 'CDS'] = np.nan
act_ca.loc[act_ca['CCode']==0, 'CCode'] = np.nan
act_ca.loc[act_ca['CDCode']==0, 'CDCode'] = np.nan
act_ca.loc[act_ca['SCode']==0, 'SCode'] = np.nan
act_ca.loc[act_ca['SName']==0, 'SName'] = np.nan
act_ca.loc[act_ca['DName']==0, 'DName'] = np.nan

#check null data for data set. 
print(act_ca.isnull().sum()) 
#CDS, CCode, CDCode and SCode has an extra null value compared to SName, DName & CName due to
#the overall State of California value.
```

### 2. Apply standard deviation

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
#numbers being a list of numbers
numbers = [1, 2, 4, 5, 6, 10]

#lambda function - mean of list of numbers
num_mean = lambda num:  sum(num) / len(num) 

print(num_mean(numbers))
```

```python
#Create a function to measure standard deviation
def num_stdev(list): 
    n = len(list)
    u = num_mean(list)
    var = sum((num-u)**2 for num in list) / n #for every number in list, subtract mean, square it and sum all squared numbers.
    stdev = round(var**0.5,4)                 #to make it neater, round() to 4 d.p.
    return stdev

#to check answer
#import statistics
#statistics.stdev(numbers)
```

### 1.2 Histograms

```python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels):
    nrows = int(np.ceil(len(list_of_columns)/2))         # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2,         
                           figsize=(20,18), sharey=True) 
    ax = ax.ravel()                                      # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns):         # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column], bins=20, color='C6') 
        ax[i].set_title(list_of_titles[i], fontsize=18)
        ax[i].set_xlabel(list_of_xlabels[i], fontsize=12)
        ax[i].set_ylabel('Frequency', fontsize=12)
```

```python
participation_col = ['act_participation', 'sat_participation','pct_above21_act', 'pct_tot_benchmark_sat']
titles = ['ACT Participation Rate 2019', 'SAT Participation Rate 2019','ACT scores above 21', 'SAT scores above benchmark']
participation_xlabel = ['Participation Rate (%)', 'Participation Rate (%)','ACT scores above 21 (%)', 'SAT scores above both benchmarks (%)']
subplot_histograms(new_combinedf, participation_col, titles, participation_xlabel)
```

The histogram for ACT participation rate is positive skewed.
The histogram for SAT participation rate is bell shaped and slightly positively skewed.
Both means are greater than the medians, hence the positively skewed curves.

We hope to shift the mean (peak of curve) towards the right, to increase the overall participation rate of both tests.

Similarly, we hope to shift the mean (peak of curve) towards the right for the % of people getting good scores, to increase competitiveness of students in California.

The participation rate for SATs are approximately normally distributed.
